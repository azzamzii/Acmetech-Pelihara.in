```python
import json
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
# load json data
with open('pet.json', 'w') as pet_content:
    json.dump(data, pet_content)
```

```python
train_word = []
train_tag = []
respon = []
tag = []

# Extract training sentences, labels, and responses from the loaded data
for intent in data['pet']:
    for pattern in intent['patterns']:
        train_word.append(pattern)
        train_tag.append(intent['tag'])
    respon.append(intent['responses'])

    if intent['tag'] not in tag:
        tag.append(intent['tag'])

num_tag = len(tag)
```

```python
num_tag
```

```python
tag_encoder = LabelEncoder()
tag_encoder.fit(train_tag)
train_tag = tag_encoder.transform(train_tag)
```

```python
tag_encoder
```

```python
num_words = 2000
max_len = 25
oov_token = "<OOV>"
embedding_dim = 16

tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)
tokenizer.fit_on_texts(train_word)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(train_word)
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len)
```

```python
model = Sequential()
model.add(Embedding(num_words, embedding_dim, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_tag, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
epochs = 1000
history = model.fit(padded_sequences, np.array(train_tag), epochs=epochs)
```

```python
import matplotlib.pyplot as plt
#plotting model accuracy
plt.plot(history.history['accuracy'],label='training set accuracy')
plt.plot(history.history['loss'],label='training set loss')
plt.legend()
```

```python
import json
import numpy as np
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder

import colorama
colorama.init()
from colorama import Fore, Style, Back

import random
import pickle

# Load JSON data
with open('pet.json', 'r') as file:
    data = json.load(file)


def chat():
    # Load trained model
    model = keras.models.load_model('chat_model.h5')

    # Load tokenizer object
    with open('tokenizer.pickle', 'rb') as handle:
        tokenizer = pickle.load(handle)

    # Load label encoder object
    with open('label_encoder.pickle', 'rb') as enc:
        tag_encoder = pickle.load(enc)

    # Parameters
    max_len = 25

    while True:
        print(Fore.LIGHTBLUE_EX + "User: " + Style.RESET_ALL, end="")
        inp = input()
        if inp.lower() == "quit":
            print("Terimakasih, sampai jumpa lagi!")
            break

        result = model.predict(keras.preprocessing.sequence.pad_sequences(
            tokenizer.texts_to_sequences([inp]), truncating='post', maxlen=max_len))
        tag = tag_encoder.inverse_transform([np.argmax(result)])

        found_tag = False  # Variable to check if a valid response was found

        for i in data['pet']:
            if i['tag'] == tag:
                print(Fore.GREEN + "ChatBot:" + Style.RESET_ALL, np.random.choice(i['responses']))
                found_tag = True
                break

        if not found_tag:
            print(Fore.GREEN + "ChatBot:" + Style.RESET_ALL, "Maaf, PawBot tidak mengerti. Berikan input nama hewan dan penyakitnya, nanti PawBot akan memberi solusi penanganannya:)")

print(Fore.YELLOW + "Start messaging with the bot (type quit to stop)!" + Style.RESET_ALL)
chat()
```
